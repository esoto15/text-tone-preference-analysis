## K-Modes

```python
import matplotlib.pyplot as plt
```

```python
# Elbow curve to find optimal K 
cost = [] 
K = range(1,10) 
for k in list(K): 
    kmode = KModes(n_clusters=k, init = "random", n_init = 5, verbose=1) 
    kmode.fit_predict(df) 
    cost.append(kmode.cost_) 
      
plt.plot(K, cost, 'x-') 
plt.xlabel('No. of clusters') 
plt.ylabel('Cost') 
plt.title('Elbow Curve') 
plt.show()
```

```python
kmodes = KModes(n_clusters=5, init='random', n_init=5, verbose=1)

# Fit the K-modes model
clusters = kmodes.fit_predict(df)
```

```python
df['Cluster'] = clusters
```

```python
df
```

```python
print("Cluster centroids:\n", kmodes.cluster_centroids_)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score, adjusted_rand_score
from scipy.spatial.distance import hamming
```

```python

# Function to compute Silhouette Score using Hamming distance
def silhouette_score_hamming(data, clusters):
    hamming_distances = [[hamming(x, y) for y in data] for x in data]
    silhouette_avg = silhouette_samples(hamming_distances, clusters).mean()
    return silhouette_avg

# Compute Silhouette Score
silhouette_avg = silhouette_score_hamming(df.values, clusters)
print("Silhouette Score:", silhouette_avg)
```

## K-Prototype

**Description**:
K-prototype is a clustering algorithm that combines k-means with k-modes to handle datasets with both numerical and categorical data. It assigns weights to both types of features and calculates a weighted distance metric to determine cluster assignments.

```python
pro_data = pd.read_excel('Datasets/Dataset_K_prototypes.xlsx')
```

```python
pro_data.drop('disability', axis=1, inplace=True)
```

```python
pro_data
```

```python
from kmodes.kprototypes import KPrototypes
```

```python
categorical_indices = [1, 2, 3, 4, 5, 7, 8, 9, 10]

def calculate_cost(data, categorical_features, max_clusters=10):
    costs = []
    for k in range(1, max_clusters + 1):
        kproto = KPrototypes(n_clusters=k, init='Cao', verbose=0)
        kproto.fit_predict(data, categorical=categorical_features)
        costs.append(kproto.cost_)
    return costs

# Calculate costs for different numbers of clusters
max_clusters = 10
costs = calculate_cost(pro_data, categorical_indices, max_clusters)

# Plot the elbow plot
plt.plot(range(1, max_clusters + 1), costs, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Cost')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
kproto = KPrototypes(n_clusters=3, init='Cao', verbose=2)

# Fit the model to your data
clusters = kproto.fit_predict(pro_data, categorical=[1,2,3,4,5,7,8,9])  # Specify indices of categorical features
```

```python
# Cluster centroids
print(kproto.cluster_centroids_)

# Cluster labels
print(kproto.labels_)
```
