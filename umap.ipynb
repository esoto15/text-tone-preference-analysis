```python
import pandas as pd
import umap
import matplotlib.pyplot as plt
import numpy as np
```

```python
encoded_data = pd.read_excel('Datasets/Encoded_Data_for_Model_Training_long.xlsx')
```

```python
umap_model = umap.UMAP(n_components=2, metric='euclidean', n_neighbors=15)
```

```python
umap_result = umap_model.fit_transform(encoded_data)
```

```python
labels = np.random.randint(0, 2, size=len(umap_result))  # Generate random binary labels

plt.scatter(umap_result[:, 0], umap_result[:, 1], c=labels, cmap='viridis')
plt.title('UMAP Visualization')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.colorbar()
plt.show()
```
