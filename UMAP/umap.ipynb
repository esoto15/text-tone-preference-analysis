```python
encoded_data = pd.read_excel('C:/Users/Owner/Desktop/research-machine-learning/Datasets/Encoded_Data_for_Model_Training_long.xlsx')
```

```python
reducer = umap.UMAP(n_components=3, metric='euclidean', n_neighbors=10, min_dist=0.1)
```

```python
embedding = reducer.fit_transform(encoded_data)
embedding.shape
```

```python
fig = px.scatter(x=embedding[:,0], y=embedding[:,1], labels={'x': 'UMAP Dimension 1', 'y': 'UMAP Dimension 2'})
fig.show()
```
